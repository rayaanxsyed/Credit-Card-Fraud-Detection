```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
x = data.drop(['Class'],axis = 1)
y = data['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
```

```python
classifier = {
    'LogisticRegression' : LogisticRegression(),
    'DecisionTreeClassifier' : DecisionTreeClassifier()
}

for name, clf in classifier.items():
    print(f"\n======{name}=======")
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    accuracy_score(y_test,y_pred)
    print(f"Accuracy Score : {accuracy_score(y_test,y_pred)}")
    print(f"Precision : {precision_score(y_test,y_pred)}")
    print(f"Recall : {recall_score(y_test,y_pred)}")
    print(f"F1 Score : {f1_score(y_test,y_pred)}")
```

```python
# Undersampling to work with unbalanced dataset

normal = data[data['Class'] == 0]
fraud = data[data['Class'] == 1]
normal.shape
```

```python
fraud.shape
```

```python
normal_sample = normal.sample(n=473) # Sample 473 rows so we can balance the dataset through undersampling
```

```python
normal_sample.shape
```

```python
new_data = pd.concat([normal_sample,fraud], ignore_index = True) # Combine the data
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
x = new_data.drop(['Class'],axis = 1)
y = new_data['Class']
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
classifier = {
    'LogisticRegression' : LogisticRegression(),
    'DecisionTreeClassifier' : DecisionTreeClassifier()
}

for name, clf in classifier.items():
    print(f"\n======{name}=======")
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    accuracy_score(y_test,y_pred)
    print(f"Accuracy Score : {accuracy_score(y_test,y_pred)}")
    print(f"Precision : {precision_score(y_test,y_pred)}")
    print(f"Recall : {recall_score(y_test,y_pred)}")
    print(f"F1 Score : {f1_score(y_test,y_pred)}")
```

```python
# Oversampling

x = data.drop(['Class'],axis = 1)
y = data['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x_res, y_res =SMOTE().fit_resample(x,y)
```

```python
y_res.value_counts()
```

```python
x_train, x_test,y_train,y_test = train_test_split(x_res,y_res, test_size = 0.2, random_state = 42)
```

```python
classifier = {
    'LogisticRegression' : LogisticRegression(),
    'DecisionTreeClassifier' : DecisionTreeClassifier()
}

for name, clf in classifier.items():
    print(f"\n======{name}=======")
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    accuracy_score(y_test,y_pred)
    print(f"Accuracy Score : {accuracy_score(y_test,y_pred)}")
    print(f"Precision : {precision_score(y_test,y_pred)}")
    print(f"Recall : {recall_score(y_test,y_pred)}")
    print(f"F1 Score : {f1_score(y_test,y_pred)}")
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(x_res,y_res)
```

```python
import joblib
```

```python
joblib.dump(dtc, "credit_card_model.pkl")
```
